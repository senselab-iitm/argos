<h1>Importing the Libraries</h1>

```python
import sys
sys.path.append("../")

from sionna.rt import load_scene
from app.configs import *
import os
```

<h1>Initializing an Iterator for iterating through the objects</h1>

```python
iterator = 0
iter_lim = None
```

<h1>Iterating through the objects</h1>

Execute this code cell as many times as you would like. An object would be highlighted in bright red and its name would be displayed with the text "Highlighting: " printed before it. In case of a custom twin should you need to hide the bounding box or floor for proper visibility, or for any other debugging purpose, this tool will prove to be helpful in noting down object identifiers.

```python
scene=None
try:
    files = os.listdir(cfg_mesh_directory)
except FileNotFoundError:
    raise Exception('Directory not found! Check configs.py and put proper cfg_mesh_directory')
xml_file = None
for f in files:
    if f.endswith('.xml'):
        xml_file = f
        print(f"XML file found: {cfg_mesh_directory}/{f}")
        scene = load_scene(f"{cfg_mesh_directory}/{f}")
        print("Scene loaded!")

        break
if xml_file is None:
    raise Exception('No xml file found! Check configs.py and put proper cfg_mesh_directory')

iter_lim = len(list(scene.objects.keys()))
scene.get(list(scene.objects.keys())[iterator]).radio_material.color=(1,0,0)
print("Highlighting: ",list(scene.objects.keys())[iterator])
scene.preview(clip_at=cfg_clip_at, clip_plane_orientation=cfg_clip_plane_orientation)
iterator=(iterator+1)%iter_lim
```
